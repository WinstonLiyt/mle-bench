```python
import matplotlib.pyplot as plt

x_values = []
y_values = []

with open('/home/v-yuanteli/mle-bench/show/w_submission_each_loop/2024-11-11T10-58-58-GMT_run-group_aide/x_score_mapping.txt', 'r') as file:
    for line in file:
        x, y = line.split()
        x_values.append(int(x))
        y_values.append(float(y))

sorted_data = sorted(zip(x_values, y_values))
x_values, y_values = zip(*sorted_data)

plt.plot(x_values, y_values, marker='o')

plt.title('spaceship-titanic aide score')
plt.xlabel('loop')
plt.ylabel('accuracy')
plt.grid(True)

plt.show()
```
